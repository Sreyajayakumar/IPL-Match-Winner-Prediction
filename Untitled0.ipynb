```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# =========================================
# STEP 1: CREATE IPL DATASET (MANUAL)
# =========================================

data = {
    "team1": ["CSK","MI","RCB","KKR","CSK","MI","RCB","KKR","CSK","MI"],
    "team2": ["MI","CSK","KKR","RCB","KKR","RCB","CSK","MI","RCB","KKR"],
    "toss_winner": ["MI","CSK","RCB","KKR","CSK","MI","RCB","KKR","CSK","MI"],
    "toss_decision": ["bat","field","bat","field","bat","field","bat","field","bat","field"],
    "venue": ["Wankhede","Chepauk","Chinnaswamy","Eden","Chepauk","Wankhede","Chinnaswamy","Eden","Chepauk","Wankhede"],
    "winner": ["MI","CSK","RCB","KKR","CSK","MI","RCB","KKR","CSK","MI"]
}

df = pd.DataFrame(data)

print("Original Dataset:\n")
print(df)

# =========================================
# STEP 2: SELECT REQUIRED COLUMNS
# =========================================

df = df[['team1','team2','toss_winner','toss_decision','venue','winner']]

print("\nDataset After Selection:\n", df)

# =========================================
# STEP 3: HANDLE MISSING VALUES
# =========================================

df_cleaned = df.fillna(method='ffill')

print("\nMissing Values After Cleaning:\n")
print(df_cleaned.isnull().sum())

# =========================================
# STEP 4: DATA FILTERING
# =========================================

top_matches = df_cleaned[df_cleaned["winner"] == "CSK"]

print("\nMatches Won by CSK:\n")
print(top_matches)

# =========================================
# STEP 5: STATISTICAL ANALYSIS
# =========================================

print("\nWinner Count:\n", df_cleaned['winner'].value_counts())

# =========================================
# STEP 6: VISUALIZATION
# =========================================

df_cleaned['winner'].value_counts().plot(kind='bar')
plt.title("Match Wins by Teams")
plt.show()

# =========================================
# STEP 7: FEATURE ENGINEERING
# =========================================

df_cleaned['toss_win_match_win'] = (df_cleaned['toss_winner'] == df_cleaned['winner']).astype(int)

# =========================================
# STEP 8: ENCODING
# =========================================

encoders = {}
df_encoded = df_cleaned.copy()

for col in df_encoded.columns:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    encoders[col] = le

# =========================================
# STEP 9: TRAIN TEST SPLIT
# =========================================

X = df_encoded.drop('winner', axis=1)
y = df_encoded['winner']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# =========================================
# STEP 10: MODEL
# =========================================

model = RandomForestClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("\nModel Results:")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# =========================================
# STEP 11: PREDICTION
# =========================================

def predict(team1, team2, toss_winner, toss_decision, venue):
    toss_feature = 1 if toss_winner == team1 else 0

    input_data = [[
        encoders['team1'].transform([team1])[0],
        encoders['team2'].transform([team2])[0],
        encoders['toss_winner'].transform([toss_winner])[0],
        encoders['toss_decision'].transform([toss_decision])[0],
        encoders['venue'].transform([venue])[0],
        toss_feature
    ]]

    pred = model.predict(input_data)
    return encoders['winner'].inverse_transform(pred)[0]

print("\nPrediction:", predict("CSK","MI","MI","bat","Wankhede"))

# =========================================
# STEP 12: NUMPY OPERATIONS
# =========================================

arr = np.random.randint(10,100,12)

print("\nArray:", arr)
print("Mean:", np.mean(arr))
print("Sum:", np.sum(arr))
print("Product:", np.prod(arr))

reshaped = arr.reshape(12,1)
print("\nReshaped:\n", reshaped)

print("\nTranspose:\n", reshaped.T)

# =========================================
# STEP 13: NaN HANDLING (NUMPY)
# =========================================

arr2 = np.array([
    [10, 20, np.nan],
    [30, np.nan, 50],
    [60, 70, 80]
])

print("\nOriginal Array:\n", arr2)

col_mean = np.nanmean(arr2, axis=0)
inds = np.where(np.isnan(arr2))
arr2[inds] = np.take(col_mean, inds[1])

print("\nAfter Replacing NaN:\n", arr2)

# =========================================
# STEP 14: VISUAL ANALYSIS
# =========================================

sns.countplot(x='toss_decision', data=df_cleaned)
plt.title("Toss Decision Distribution")
plt.show()
```
